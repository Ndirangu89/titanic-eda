```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For better visuals
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("/Users/joshndirangu/Documents/titanic-eda/data/cleaned_titanic.csv")
```

# DATA VISUALIZATION OF THE CLEANED DATA

 # STEP 1: Univariate Analysis

```python
df.info()
```

```python
# Visualize Survived Distribution
ap = sns.countplot(x="Survived", data = df)
plt.title("Survival Count [0 = No, 1 = Yes]")

# Add counts on top of each bar
for container in ap.containers:
    ap.bar_label(container, label_type='edge')

# Saving the diagram to my images subfolder
plt.savefig('/Users/joshndirangu/Documents/titanic-eda/images/survival_Dist.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
# Visualize Sex
ax = sns.countplot(x="Sex", data = df)
plt.title("Gender Count Visual")

# Add counts on top of each bar
for container in ax.containers:
    ax.bar_label(container, label_type='edge')

# Saving the diagram to my images subfolder
plt.savefig('/Users/joshndirangu/Documents/titanic-eda/images/gender_Visual.png', dpi=300, bbox_inches='tight')
    
plt.show()
```

```python
# Visualize Passenger Classes
ax = sns.countplot(x="Pclass", data = df)
plt.title("Passenger Class Visual")

# Add counts on top of each bar
for container in ax.containers:
    ax.bar_label(container, label_type='edge')

# Saving the diagram to my images subfolder
plt.savefig('/Users/joshndirangu/Documents/titanic-eda/images/Pclass_Visual', dpi=300, bbox_inches='tight')
    
plt.show()
```

BIVARIATE ANALYSIS

```python
# Sex vs Survival

ap = sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival Count by Gender')

# Add counts on top of each bar
for container in ap.containers:
    ap.bar_label(container, label_type='edge')

# Saving the diagram to my images subfolder
plt.savefig('/Users/joshndirangu/Documents/titanic-eda/images/sex_Vs_Surv.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
# Survival vs Passenger Class

ax = sns.countplot(x='Pclass', hue='Survived', data=df)

plt.title('Survival Count by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')

# Add count labels on each bar
for container in ax.containers:
    ax.bar_label(container, label_type='edge')

# Saving the diagram to my images subfolder
plt.savefig('/Users/joshndirangu/Documents/titanic-eda/images/surv_VS_pass.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
# Countplot: Visual of Embarked vs Survival

ap = sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title("Survival by Embarkation Point")
plt.xlabel("Embarkation Points")
plt.ylabel("Passenger Count")

#Adding count above each bar
for container in ap.containers:
    ap.bar_label(container, label_type='edge')

# Saving the diagram to my images subfolder
plt.savefig('/Users/joshndirangu/Documents/titanic-eda/images/survi_VS_embark.png', dpi=300, bbox_inches='tight')
    
plt.show()
```

```python
# Box Plot showing Age vs Survival
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age Distribution by Survival')

# Saving the diagram to my images subfolder
plt.savefig('/Users/joshndirangu/Documents/titanic-eda/images/age_VS_survBplot.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# KDE Plot, Survival vs Gender. Why use KernelDensityEstimate plot?

g = sns.FacetGrid(df, hue='Survived', col='Sex', height=5)
g.map(sns.kdeplot, 'Age', fill=True).add_legend()
g.fig.suptitle("Age Distribution by Survival and Gender", y=1.05)

# Saving the diagram to my images subfolder
plt.savefig('/Users/joshndirangu/Documents/titanic-eda/images/survival_VS_genderKDE.png', dpi=300, bbox_inches='tight')

plt.show()
```

MULTIVARIATE ANALYSIS


```python
# Shows the correlation of each column to each of the other columns

corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')

# Saving the diagram to my images subfolder
plt.savefig('/Users/joshndirangu/Documents/titanic-eda/images/corr.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
# Way to visualize multiple relationships at once visually

sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived')
plt.suptitle("Pair Plot of Key Variables", y=1.02)

# Saving the diagram to my images subfolder
plt.savefig('/Users/joshndirangu/Documents/titanic-eda/images/gen_visual.png', dpi=300, bbox_inches='tight')

plt.show()
```
